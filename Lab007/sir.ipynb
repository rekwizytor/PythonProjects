```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def deriv(y, t, N, beta, gamma):
    tmp_S, tmp_I, tmp_R = y
    dSdt = -beta * tmp_S * tmp_I / N
    dIdt = beta * tmp_S * tmp_I / N - gamma * tmp_I
    dRdt = gamma * tmp_I
    return dSdt, dIdt, dRdt
```

```python
time = np.linspace(0, 100, 100)
```

```python
N = 100 # Population size
I0 = 5 # Initial number of infected individuals
R0 = 1 # Initial number of recovered individuals
S0 = N - I0 - R0 # Initial number of susceptible individuals
beta = 0.2 # Infection rate
gamma = 0.1 # Recovery rate

y0 = S0, I0, R0 # Initial conditions vector
S, I, R = odeint(deriv, y0, time, args=(N, beta, gamma)).T

plt.figure(figsize=(6, 4), dpi=300)
plt.plot(time, S, label='S(t)')
plt.plot(time, I, label='I(t)')
plt.plot(time, R, label='R(t)')
plt.legend(loc='best')
plt.xlabel('time (days)')
plt.ylim(-5, 105)
plt.ylabel('SIR population')
plt.title(f'Model SIR: N={N}, I$_0$={I0}, R$_0$={R0}, $\\beta=${beta}, $\\gamma$={gamma}')
plt.tight_layout()
plt.savefig('SIR_1.pdf')
plt.show()
```
