```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

Oscylator harmoniczny z tłumieniem $$\frac{d^2x}{dt^2} = - \omega_0^2 x - 2\gamma \frac{dx}{dt} $$

```python
x = sp.Function('x')
t = sp.Symbol('t', real=True)
gamma = sp.Symbol('gamma', positive=True)
w0 = sp.Symbol('omega_0', positive=True)
x0 = sp.Symbol('x_0', real=True)
v0 = sp.Symbol('v_0', real=True)

x(t), t, gamma, w0, x0, v0
```

```python
eq = sp.Eq(x(t).diff(t, t), - 2 * gamma * x(t).diff(t) - w0**2 * x(t))
eq
```

```python
sol = sp.dsolve(eq, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
```

```python
solf1 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, gamma: 1, x0: 1.0, v0: 1.0}))
solf2 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, gamma: 5.1, x0: 1.0, v0: 1.0}))
solf3 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, gamma: 10, x0: 1.0, v0: 1.0}))
```

```python
time = np.linspace(0, 10, 1000)
plt.figure(figsize=(7, 4), dpi=300)
plt.plot(time, solf1(time), label='$\gamma$ = 1.0, $\omega_0$ = 5.0')
plt.plot(time, solf2(time), label='$\gamma$ = 5.1, $\omega_0$ = 5.0')
plt.plot(time, solf3(time), label='$\gamma$ = 10.0, $\omega_0$ = 5.0')
plt.xlabel('t')
plt.ylabel('$x(t)$')
plt.grid(linestyle='--')
plt.legend()
plt.tight_layout()
plt.show()
```

Oscylator harmoniczny z wymuszeniem $$\frac{d^2x}{dt^2} = - \omega_0^2 x + \alpha_0 cos(\omega t) $$

```python
x = sp.Function('x')
t = sp.Symbol('t', real=True)
w0 = sp.Symbol('omega_0', positive=True)
w = sp.Symbol('omega', positive=True)
alpha0 = sp.Symbol('alpha_0', positive=True)
x0 = sp.Symbol('x_0', real=True)
v0 = sp.Symbol('v_0', real=True)

x(t), t, w, w0, alpha0, x0, v0
```

```python
eq = sp.Eq(x(t).diff(t, t), - w0**2 * x(t) + alpha0 * sp.cos(w * t))
eq
```

```python
sol = sp.dsolve(eq, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol
```

```python
solf1 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 0, alpha0: 0, x0: 1.0, v0: 1.0}))
solf2 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 1, alpha0: 5, x0: 1.0, v0: 1.0}))
solf3 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 5.1, alpha0: 5, x0: 1.0, v0: 1.0}))
solf4 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 7, alpha0: 20, x0: 1.0, v0: 1.0}))
```

```python
time = np.linspace(0, 20, 1000)
plt.figure(figsize=(7, 4), dpi=300)

plt.subplot(221)
plt.plot(time, solf1(time))
plt.title('$\omega_0 = 5.0, \omega = 0, \\alpha_0 = 0$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(222)
plt.plot(time, solf2(time), color='orange')
plt.title('$\omega_0 = 5.0, \omega = 1, \\alpha_0 = 5.0$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(223)
plt.plot(time, solf3(time), color='green')
plt.title('$\omega_0 = 5.0, \omega = 5.1, \\alpha_0 = 5.0$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(224)
plt.plot(time, solf4(time), color='purple')
plt.title('$\omega_0 = 5.0, \omega = 7.0, \\alpha_0 = 20.0$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.tight_layout()
```

Oscylator harmoniczny z tłumieniem i wymuszeniem $$\frac{d^2x}{dt^2} = - \omega_0^2 x - 2\gamma \frac{dx}{dt} + \alpha_0 cos(\omega t) $$

```python
x = sp.Function('x')
t = sp.Symbol('t', real=True)
w0 = sp.Symbol('omega_0', positive=True)
w = sp.Symbol('omega', positive=True)
gamma = sp.Symbol('gamma', positive=True)
alpha0 = sp.Symbol('alpha_0', positive=True)
x0 = sp.Symbol('x_0', real=True)
v0 = sp.Symbol('v_0', real=True)

x(t), t, w, w0, gamma, alpha0, x0, v0
```

```python
eq = sp.Eq(x(t).diff(t, t), - w0**2 * x(t) - 2 * gamma * x(t).diff(t) + alpha0 * sp.cos(w * t))
eq
```

```python
sol = sp.dsolve(eq, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol.simplify()
```

```python
solf1 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 2, gamma: 1, alpha0: 2, x0: 1.0, v0: 1.0}))
solf2 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 2, gamma: 4, alpha0: 2, x0: 1.0, v0: 1.0}))
solf3 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 15, gamma: 1, alpha0: 10, x0: 1.0, v0: 1.0}))
solf4 = sp.lambdify(t, sol.rhs.subs({w0: 5.0, w: 15, gamma: 0.05, alpha0: 10, x0: 1.0, v0: 1.0}))
```

```python
time = np.linspace(0, 20, 1000)
plt.figure(figsize=(7, 4), dpi=300)

plt.subplot(221)
plt.plot(time, solf1(time))
plt.title('$\omega_0 = 5.0, \gamma = 1, \omega = 2, \\alpha_0 = 2$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(222)
plt.plot(time, solf2(time), color='orange')
plt.title('$\omega_0 = 5.0, \gamma = 4, \omega = 2, \\alpha_0 = 2$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(223)
plt.plot(time, solf3(time), color='green')
plt.title('$\omega_0 = 5.0, \gamma = 1, \omega = 15, \\alpha_0 = 10$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.subplot(224)
plt.plot(time, solf4(time), color='purple')
plt.title('$\omega_0 = 5.0, \gamma = 0.05, \omega = 15, \\alpha_0 = 10.$')
plt.xlabel('t')
plt.ylabel('$x(t)$')

plt.tight_layout()
```
